Importing the Modules

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

Reading the dataset from kaggle

```python
data = pd.read_csv("data.csv")
data.head(10) #Reading the first 10 values
```

Removing the missing values if any 

```python
data.isnull().any()
missing_values = ['NONE', 'None', 'none', 'NAN', 'Nan', 'nan', 'NA', 'Na', 'na']
data = pd.read_csv("data.csv",na_values=missing_values)
data.isnull().sum()

#removing the missing values
data.dropna(inplace=True)
data.info()
```

Spliting the data into 80/20 for training and testing

```python
from sklearn.model_selection import train_test_split

x = data.drop("Bankrupt?",axis = 1)
y = data["Bankrupt?"]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

print(y_train)
```
